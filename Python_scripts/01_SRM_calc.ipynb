```python
#%% Dataframe Adjustments
pd.set_option('display.max_columns', 500)

# delete last 5 rows
NDC = NDC[:-5]


#turn 'object' to 'float'
NDC.iloc[:,1:24] = NDC.iloc[:,1:24].astype(float)

#extend df until 2500
val_2300 = list(range(2101,2501))

#for ye in val_2300:
#    NDC = NDC.append({'YEARS' : str(ye)}, ignore_index = True)

NDC_p = NDC.copy()
NDC_10 = NDC.copy()
NDC_5 = NDC.copy()

SSP1_19 = SSP1_19.rename(columns={"HFC125": "HFC-125", "HFC23": "HFC-23", "HFC32": "HFC-32"})
```

```python
#%% absolute decline calc
    
        
def min_values(val, rate, gas):
    if math.isclose(val, min(SSP1_19[gas]), abs_tol=0.03):
        return min(SSP1_19[gas])
    elif val <= min(SSP1_19[gas]):
        return min(SSP1_19[gas])
    else: 
        return val - rate

for ye in val_2300:
    NDC = NDC.append({'YEARS' : str(ye)}, ignore_index = True)
    for col in NDC.columns[1:]:  
        val_2081 = np.float32(NDC.loc[(NDC.YEARS == '2081'), col].values).item()
        val_2100 = np.float32(NDC.loc[(NDC.YEARS == '2100'), col].values).item()
        val_ye = np.float32(NDC.loc[(NDC.YEARS == str(ye-1)), col].values).item()
        #absolute decline per year
        abs_decl = (val_2081 - val_2100) / 20
        #Dec = end_zero(val_ye, abs_decl)
        Dec = min_values(val_ye, abs_decl, col)
        
        NDC.loc[(NDC.YEARS == str(ye)), col] = Dec

# Define end of CDR   
ye_after_SRM = list(range(2284,2501))  #2318 !!!

for ye in ye_after_SRM:
    NDC.loc[(NDC.YEARS == str(ye)), 'FossilCO2'] = 0 



# calc with 10 neg CO2

def min_values_10(val, rate, gas):   
    if gas == 'FossilCO2':
        if math.isclose(val, -2.725, abs_tol=0.03):
            return -2.725
        elif val <= -2.725:   #10000 Mt CO2 = 2,7 GtC
            return -2.725
        else:
            return val - rate
    #elif val <= 0:
    #    return 0
    else:
        if math.isclose(val, min(SSP1_19[gas]), abs_tol=0.03):
            return min(SSP1_19[gas])
        elif val <= min(SSP1_19[gas]):
            return min(SSP1_19[gas])
        else: 
            return val - rate
       # else:
        #    return val - rate


for ye in val_2300:
    NDC_10 = NDC_10.append({'YEARS' : str(ye)}, ignore_index = True)
    for col in NDC_10.columns[1:]:  
        val_2081 = np.float32(NDC_10.loc[(NDC_10.YEARS == '2081'), col].values).item()
        val_2100 = np.float32(NDC_10.loc[(NDC_10.YEARS == '2100'), col].values).item()
        val_ye = np.float32(NDC_10.loc[(NDC_10.YEARS == str(ye-1)), col].values).item()
        #absolute decline per year
        abs_decl = (val_2081 - val_2100) / 20
        #Dec = end_zero(val_ye, abs_decl)
        Dec = min_values_10(val_ye, abs_decl, col)
        
        NDC_10.loc[(NDC_10.YEARS == str(ye)), col] = Dec

# Define end of CDR   
ye_after_SRM = list(range(2292,2501))  ##2331 !!!

for ye in ye_after_SRM:
    NDC_10.loc[(NDC_10.YEARS == str(ye)), 'FossilCO2'] = 0
    


# calc with 5 neg CO2

        
def min_values_5(val, rate, gas):
    if gas == 'FossilCO2':
        if math.isclose(val, -1.362, abs_tol=0.03):
            return -1.362
        elif val <= -1.362:   #5000 MtCO2 = 1,362 GtC
            return -1.362
        else:
            return val - rate
    #elif val <= 0:
    #    return 0
    else:
        if math.isclose(val, min(SSP1_19[gas]), abs_tol=0.03):
            return min(SSP1_19[gas])
        elif val <= min(SSP1_19[gas]):
            return min(SSP1_19[gas])
        else: 
            return val - rate
       # else:
        #    return val - rate


for ye in val_2300:
    NDC_5 = NDC_5.append({'YEARS' : str(ye)}, ignore_index = True)
    for col in NDC_5.columns[1:]:  
        val_2081 = np.float32(NDC_5.loc[(NDC_5.YEARS == '2081'), col].values).item()
        val_2100 = np.float32(NDC_5.loc[(NDC_5.YEARS == '2100'), col].values).item()
        val_ye = np.float32(NDC_5.loc[(NDC_5.YEARS == str(ye-1)), col].values).item()

        #absolute decline per year
        abs_decl = (val_2081 - val_2100) / 20
        #Dec = end_zero(val_ye, abs_decl)
        Dec = min_values_5(val_ye,  abs_decl, col)
        
        NDC_5.loc[(NDC_5.YEARS == str(ye)), col] = Dec

# Define end of CDR   
ye_after_SRM = list(range(2354,2501))    #2400 

for ye in ye_after_SRM:
    NDC_5.loc[(NDC_5.YEARS == str(ye)), 'FossilCO2'] = 0   
    #0.65GtC to remove the forcing from non-CO2 gases
```


## SRM calculation with carbon budget 


```python
#%% CO2eq transformation

# Convert all units to Mt

def conf_unit(df):
    for col in df.columns[12:]:
        df[col] = df[col]/1000
    for col in df.columns[1:3]:
        df[col] = df[col]*1000*3.67  #convert GtC to MtCO2
    return df
        
NDC = conf_unit(NDC)
NDC_5 = conf_unit(NDC_5)
NDC_10 = conf_unit(NDC_10)
#SSP1_19 = conf_unit(SSP1_19)

GWP_dict = {
        'CH4': 28,      # 34  -> mit cc fb
        'N2O': 265,     # 298 -> mit cc fb
        'SOx': 0,       ### ??
        'CO': 2.2, 
        'NMVOC': 4.5,
        'NOx': -11,
        'BC': 460,
        'OC': 55,
        'NH3': 0,       # ??
        'CF4': 6630,    # 7350 -> mit cc fb
        'C2F6': 11100,
        'C6F14': 7910,
        'HFC-23': 12400,
        'HFC-32': 677,
        'HFC43-10': 1650,
        'HFC-125': 3170,
        'HFC134a': 1300, # 1550 -> mit cc fb
        'HFC143a': 4800,
        'HFC227ea': 3350,
        'HFC245fa': 858,
        'SF6': 23500}


# multiply by GWP

def conf_CO2eq(df):
    for col in df.columns[3:]:
        df[col] = df[col]*GWP_dict[col]
    return df

NDC_eq = conf_CO2eq(NDC)
NDC_5 = conf_CO2eq(NDC_5)
NDC_10 = conf_CO2eq(NDC_10)
```

```python
#%% create sum column

def sum_column(df):
    df['sum_col'] = df.iloc[:,1:].sum(axis=1) #df[1:].sum(axis=1)
    return df

NDC_eq = sum_column(NDC_eq)
NDC_5 = sum_column(NDC_5)
NDC_10 = sum_column(NDC_10)


#%% cum em column

def cum_column(df):
    df['cum_em'] = df['sum_col'].cumsum()
    return df
    
NDC_eq = cum_column(NDC_eq)
NDC_5 = cum_column(NDC_5)
NDC_10 = cum_column(NDC_10)



#%% separate cum columns

def sum_column(df):
    df['sum_col_co2'] = df.iloc[:,1:3].sum(axis=1) 
    df['sum_col_nonco2'] = df.iloc[:,3:24].sum(axis=1) 
    df['cum_em_co2'] = df['sum_col_co2'].cumsum()
    df['cum_em_nonco2'] = df['sum_col_nonco2'].cumsum()
    return df

NDC_eq = sum_column(NDC_eq)
NDC_5 = sum_column(NDC_5)
NDC_10 = sum_column(NDC_10)


#%% total CO2

def tot_co2(df):
    df['TotalCO2'] = df['FossilCO2'] + df['OtherCO2']
    return df

NDC_eq = tot_co2(NDC_eq)
NDC_5 = tot_co2(NDC_5)
NDC_10 = tot_co2(NDC_10)
```

```python
#%% carbon budget CONSTRAIN

def SRM_start_end(df):
    df_start = df[:100]
    df_end = df[230:]
    SRM_start = df_start.iloc[(df_start.cum_em_co2 - 985000).abs().argsort()[:1]]
    SRM_end = df_end.iloc[(df_end.cum_em_co2 - 985000).abs().argsort()[:1]]
    SRM_start = SRM_start.YEARS
    SRM_end = SRM_end.YEARS
    SRM = [SRM_start.values.item(), SRM_end.values.item()]
    return SRM


SRM_eq = SRM_start_end(NDC_eq)
SRM_5 = SRM_start_end(NDC_5)
SRM_10 = SRM_start_end(NDC_10)
```

```python
#%% carbon budget CONSTRAIN

def SRM_start_end_15(df):
    df_start = df[:100]
    df_end = df[390:]
    SRM_start = df_start.iloc[(df_start.cum_em_co2 - 235000).abs().argsort()[:1]]
    SRM_end = df_end.iloc[(df_end.cum_em_co2 - 235000).abs().argsort()[:1]]
    SRM_start = SRM_start.YEARS
    SRM_end = SRM_end.YEARS
    SRM = [SRM_start.values.item(), SRM_end.values.item()]
    return SRM


SRM_15_eq = SRM_start_end_15(NDC_eq)
SRM_15_5 = SRM_start_end_15(NDC_5)
SRM_15_10 = SRM_start_end_15(NDC_10)
```

```python
# CDR quantification


def CDR_quantification(df):
    df['CDR'] = 0
    for ye in df.YEARS:
        df.loc[(df.YEARS == ye)& (df.FossilCO2 < 0), 'CDR'] = df.loc[(df.YEARS == ye) & (df.FossilCO2 < 0), 'FossilCO2'].abs()
    return df


NDC_eq = CDR_quantification(NDC_eq)
NDC_5 = CDR_quantification(NDC_5)
NDC_10 = CDR_quantification(NDC_10)

def CDR_sum(df): #CDR total sum in Mt Co2    
    Index_label = df.query('YEARS < "2500"').index.tolist()
    CDR_sum = df.loc[Index_label,'CDR'].sum()
    return CDR_sum
    
CDR_eq = CDR_sum(NDC_eq) 
CDR_5 = CDR_sum(NDC_5) 
CDR_10 = CDR_sum(NDC_10)  
```
